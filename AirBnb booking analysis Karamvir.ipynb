## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# make dataset copy
df1 = df.copy()
# Use groupby method to count property booked location wise
df1.groupby('neighbourhood_group').size()
````

````python
# find out type of room booked
df1.groupby('room_type').size()
````

````python
# find out price of property locationwise
df1.groupby('neighbourhood_group')['price'].mean()
````

````python
# find out price of property according to roomtype
df1.groupby('room_type')['price'].mean()
````

````python
# Availability of room as per location
df1.groupby('neighbourhood_group')['availability_365'].mean()
````

````python
# find out price of property according to availability
df1.groupby('availability_365')['price'].mean().sort_values(ascending = False)
````

````python
# calculate the property name that booked maximum time
df1.groupby('name').size()
````

````python
# calculate the price according to minimum night criteria
df1.groupby('minimum_nights')['price'].mean().sort_values(ascending = False)
````

````python
# calculate the number of booking according to host name
df1.groupby('host_name').size()
````

````python
# find out type of room and price according to location
df1.groupby('neighbourhood_group').agg({'room_type':'size','price':'mean'})
````

````python
# find out number of bookings on the basis of number of night condition
df1.groupby('minimum_nights').size().sort_values(ascending = False)
````

````python
# find price of room on the basis of minimum no of night booking
df1.groupby('minimum_nights')['price'].mean().sort_values(ascending = False)
````
